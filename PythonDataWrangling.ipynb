c) Finding the average volume of the two cryptocurrencies in the year 2020.

```python
# First, make the "Date" column a date type and set it as the index.
btc["Date"] = pd.to_datetime(btc["Date"])
btc = btc.set_index(["Date"])
eth["Date"] = pd.to_datetime(eth["Date"])
eth = eth.set_index(["Date"])
```

d) Finally, comparing the two cryptocurrencies. First we need to merge the two dataframes based on the date.

```python
btc_eth = pd.merge(btc.add_suffix("_BTC"), eth.add_suffix("_ETH"), left_index=True, right_index=True)
btc_eth
```

Creating a column showing the difference in closing prices, also calculating and finding the averge, min, max difference.

```python
btc_eth["Close_diff"] = btc_eth["Close_BTC"] - btc_eth["Close_ETH"]
print("Average distance:", btc_eth["Close_diff"].mean())
print("Min distance:", btc_eth["Close_diff"].min())
print("Max distance:", btc_eth["Close_diff"].max())
```

Now creating a DataFrame showing the relative increase in closing prices. and addressing the following questions:
- In how many % of days did the prices BTC and ETH increase or decrease?
- In how many % of days was the change in BTC price exceed the one of ETH?

```python
btc_eth_incr = (btc_eth - btc_eth.shift()) / btc_eth.shift()
btc_eth_incr = btc_eth_incr[["Close_BTC", "Close_ETH"]].dropna()
btc_eth_incr
```

```python
btc_up = len(btc_eth_incr[btc_eth_incr["Close_BTC"]>0].index)/len(btc_eth_incr.index)*100
btc_down = len(btc_eth_incr[btc_eth_incr["Close_BTC"]<0].index)/len(btc_eth_incr.index)*100
eth_up = len(btc_eth_incr[btc_eth_incr["Close_ETH"]>0].index)/len(btc_eth_incr.index)*100
eth_down = len(btc_eth_incr[btc_eth_incr["Close_ETH"]<0].index)/len(btc_eth_incr.index)*100
print("BTC price increased in {}% of days.".format(round(btc_up,2)))
print("BTC price decreased in {}% of days.".format(round(btc_down,2)))
print("ETH price increased in {}% of days.".format(round(eth_up,2)))
print("ETH price decreased in {}% of days.".format(round(eth_down,2)))
```

```python
btc_eth_incr["BTC_above_ETH"] = (btc_eth_incr["Close_BTC"] > btc_eth_incr["Close_ETH"])
btc_above_eth = sum(btc_eth_incr["BTC_above_ETH"])/len(btc_eth_incr.index)*100
print("BTC price change above ETH price change in {}% of days.".format(round(btc_above_eth,2)))
```
